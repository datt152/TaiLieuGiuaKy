```python
import pandas as pd 
import seaborn as sns
df = pd.read_csv('dulieuxettuyendaihoc.csv',
                 header=0,delimiter=','
                 ,encoding='utf-8')
df = df[['GT','DT','KV','KT','M1',
         'M2','M3',
         'NGAYTHI','DINHHUONGNGHENGHIEP']]


df
```

```python
df.rename(columns={'NGAYTHI':'NT',
                   'DINHHUONGNGHENGHIEP':'NGHE'}, 
                   inplace=True)

df

// doi ten cot 
```

```python
df.head(5)

// hien thi 5 dong dau tien 
```

```python
df.dropna(how='all', inplace=True)

#  loai bo nhung hang bi rong
```

```python
df.drop_duplicates(inplace=True)
df
#  loai bo nhung hang bi trung
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.heatmap(df.isna().transpose(),cmap='YlGnBu',
            cbar_kws={'label':'Dữ liệu thiếu'})
plt.savefig('missingdata.png', dpi=100)

plt.show()

ve bieu do ve du lieu thieu

```

```python
df['DT'].fillna('KINH', inplace=True)


#  khao sat 
#  dua kinh nghiem
```

```python
df['M1'].fillna(df['M1'].mean(), inplace=True)
```

```python
df['M1'].fillna(df['M1'].mean(), inplace=True)
# co nghia la thay cac gia tri NaN trong cot M1
# bang gia tri trung binh cua cot M1
df
```

```python
df['TBM'] = (df['M1']+df['M2']+df['M3'])/3
df

# them bien TBM
```

```python
df.loc[df['TBM'] < 5.0, 'XL'] = 'FAIL'

# tao cot bien xep loai co dieu kien 
```

```python
df.loc[(df['TBM'] >= 5.0) & 
       (df['TBM'] < 7.0), 'XL'] = 'FAIR'
df
```

```python
dict_map = {
    'A1':'G1', 'C':'G3', 'D1': 'G3', 'A': 'G1', 'B': 'G2'
}

df['NHOM'] = df['KT'].map(dict_map)
df
```

```python
def fplus(x,y):
    if (x == 'G1' or x == 'G2') and (y >= 4.0):
        return 1.0
    else:
        return 0.0

df['PLUS'] = list(map(fplus,df['NHOM'],df['TBM']))
df

# tao cot plus
```
