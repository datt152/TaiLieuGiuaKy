```python
import pandas as pd 
import seaborn as sns
df = pd.read_csv('dulieuxettuyendaihoc.csv',
                 header=0,delimiter=','
                 ,encoding='utf-8')
df = df[['GT','DT','KV','KT','M1',
         'M2','M3',
         'NGAYTHI','DINHHUONGNGHENGHIEP']]

# in ra du lieu co cac cot la
# GT,DT,KV,KT,M1,M2,M3,NGAYTHI,
# DINHHUONGNGHENGHIEP

df
```

```python
df.rename(columns={'NGAYTHI':'NT',
                   'DINHHUONGNGHENGHIEP':'NGHE'}, 
                   inplace=True)

df

# doi ten caccot inplace=True 
# la thay doi truc tiep tren df
```

```python
df.head(5)
# hien thị 5 dong dau tien trong file 
# chay tu 0 - 4
```

```python
df.dropna(how='all', inplace=True)
df

# loai bo nhung hang bi rong , NaN
```

```python
df.drop_duplicates(inplace=True)
df
#  loai bo nhung hang bi trung
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.heatmap(df.isna().transpose(),cmap='YlGnBu',
            cbar_kws={'label':'Dữ liệu thiếu'})
plt.savefig('missingdata.png', dpi=100)
#.transpose() DUNG DE HOAN VI, HANG THANH COT, COT THANH HANG, DE DE NHIN HON
plt.show()
```

```python
df['DT'].fillna('KINH', inplace=True)
# thay the NaN trong cot DT bang chu KINH
df
```

```python
# hien thi nhung hang co cot M1 la NaN
df[df['M1'].isna()]
```

```python
df['M1'].fillna(df['M1'].mean(), inplace=True)
# neu nhu cot M1 bi rong thi thay 
# the bang gia tri trung binh cua cot M1
```

```python
df['M2'].fillna(df['M2'].median(), inplace=True)
# neu nhu cot M2 bi rong thi thay
# the bang gia tri trung vi cua cot M2
# median() gia tri trung binh vi cua cot M2

df
```

```python
df['M3'].fillna(0, inplace=True)
df
```

```python
df['TBM'] = (df['M1']+df['M2']+df['M3'])/3

# df.loc[df['TBM'] < 5.0, 'XL'] = 'FAIL' 
# Thường được sử dụng để thêm một cột mới
# dựa trên điều kiện logic, gán giá trị
# cụ thể cho các hàng thỏa mãn điều kiện.


# df['TBM'] = (df['M1']+df['M2']+df['M3'])/3
# Thường được sử dụng để tạo một cột mới
# dựa trên phép tính toán từ các cột khác trong DataFrame.
#them truc tiep len file df cot TBM
df
```

```python
dict_map = {
    'A1':'G1', 'C':'G3', 'D1': 'G3', 'A': 'G1', 'B': 'G2'
}
df['NHOM'] = df['KT'].map(dict_map)

# co nghia la neu nhu trong cot
# KT co gia tri la A1 thi
# se thay the bang G1
df
```

```python
def fplus(x,y):
    if (x == 'G1' or x == 'G2') and (y >= 4.0):
        return 1.0
    else:
        return 0.0

df['PLUS'] = list(map(fplus,df['NHOM'],df['TBM']))



df
```

```python
df.loc[df['TBM'] < 5.0, 'XL'] = 'FAIL'

df.loc[(df['TBM'] >= 5.0) & 
       (df['TBM'] < 7.0), 'XL'] = 'FAIR'

df.loc[(df['TBM'] >= 7.0) & 
       (df['TBM'] < 9.0), 'XL'] = 'GOOD'

df.loc[df['TBM'] >= 9.0, 'XL'] = 'EXCEL'
df
```
